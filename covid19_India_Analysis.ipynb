```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
total_states = np.arange(len(df['State'])) 
```

```python
total_states
```

```python
df['Active']
```

```python
max(df['Active'])
```

```python
#Active Number of Cases based on States
```

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(9, 6), dpi=80, facecolor='w', edgecolor='k')
plt.barh(total_states,df['Active'], align='center', alpha=0.5,  
                 color=(1,0,0),  
                 edgecolor=(0.5,0.2,0.8))
    
plt.yticks(total_states, df['State'])  
plt.xlim(1,max(df['Active'])+100) 
plt.xlabel('Active Number of Cases')  
plt.title('Corona Virus Cases')  
plt.show()  
```

```python
#Death Cases based on States
```

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(9, 6), dpi=80, facecolor='w', edgecolor='k')
plt.barh(total_states,df['Deaths'], align='center', alpha=0.5,  
                 color=(0,0,1),  
                 edgecolor=(0.5,0.4,0.8)  )
    
plt.yticks(total_states, df['State'])  
plt.xlim(1,max(df['Deaths'])+10) 
plt.xlabel('Death Number of Cases')  
plt.title('Corona Virus Cases')  
plt.show()  
```
